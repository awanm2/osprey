```python
import matplotlib.pyplot as plt

# Plot 2x2x2 cube with a camera perspective on a 2-d plan

A = ( 1, 1,  1)
B = (-1, 1, 1 )
C = (1,-1,1)
D = (-1,-1,1)
E = ( 1, 1,-1 )
F = (-1, 1,-1 )
G= ( 1,-1,-1 )
H=  (-1,-1,-1 )

edges = [ [A, B], [C, D], [E, F], [G,H], [A,C], [B,D], [E,G], [F,H],[A,E], [C,G], [B,F], [D,H]]

#// Assume camera is at c=(2,3,5)
#1. subtract camera c from vertex (X,Y,Z) to get (x,y,z)
#2. divide (x,y) by z to get (u,v)â€”write as a fraction
#- Draw line between (u1,v1) and (u2,v2)

def plot_all(camera, edge_list):
  #- Convert (X,Y,Z) of both endpoints to (u,v):
  #1. subtract camera c from vertex (X,Y,Z) to get (x,y,z)
  for edge in edge_list:
    vertex_1 = edge[0]
    vertex_2 = edge[1]
    vertex_1_c = tuple(map(lambda i, j: i - j, vertex_1, camera))
    vertex_2_c = tuple(map(lambda i, j: i - j, vertex_2, camera))
    x_1_c, y_1_c, z_1_c = vertex_1_c
    x_2_c, y_2_c, z_2_c = vertex_2_c

    p1 = (x_1_c/z_1_c, y_1_c/z_1_c)
    p2 = (x_2_c/z_2_c, y_2_c/z_2_c)

    x1, y1 = p1
    x2, y2 = p2

    # Create a scatter plot
    plt.scatter([x1, x2], [y1, y2], color='red', marker='o')

    # Draw a line connecting the two points
    plt.plot([x1, x2], [y1, y2], color='blue', linestyle='-', linewidth=2)


    # Set labels for the axes
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')


    #plot_points(p1, p2)



c = (2,3,4)
plot_all(c, edges)
plt.show()


# Call the function to plot the points
```
